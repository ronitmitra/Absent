```python
import pandas as pd
import numpy as np
```

```python
data_preprocessed = pd.read_csv('/content/drive/MyDrive/Processed_Absent.csv')
data_preprocessed.head(20)
```

```python
# Create Target
median_value = data_preprocessed['Absenteeism Time in Hours'].median()
median_value
```

```python
target = np.where(data_preprocessed['Absenteeism Time in Hours']>median_value,1,0 )
target
```

```python
target.shape
```

```python
data_preprocessed['Excessive absence'] = target
data_preprocessed.head()
```

```python
df_target = data_preprocessed.drop(['Absenteeism Time in Hours'],axis=1)
df_target.head()
df_target.columns.values
```

```python
# Select Inputs
df_target.columns = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Months',
       'Day of Week', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Excessive absence']
```

```python
column_reorder = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Education','Months',
       'Day of Week', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index','Children', 'Pets', 'Excessive absence']
```

```python
df_target = df_target[column_reorder]
df_target.columns.values
```

```python
unscaled_inputs = df_target.iloc[:,5:-1]
unscaled_inputs.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = scaler.transform(unscaled_inputs)
scaled_inputs
```

```python
df_target.iloc[:,5:-1] = scaled_inputs
df_target.head(20)
```

```python
scaled_inputs_all = df_target.iloc[:,:-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(scaled_inputs_all,target,train_size=0.8,random_state=20)
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
# Creating Summary table
reg.coef_,reg.intercept_
```

```python
feature_name = scaled_inputs_all.columns.values
```

```python
summary_table = pd.DataFrame(columns = ['Feature_name'],data=feature_name)
summary_table['Coefficients'] = np.transpose(reg.coef_)
summary_table
```

```python
summary_table.index = summary_table.index + 1
summary_table.loc[0] = ['Intercept',reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

```python
summary_table['Odds_ratio'] = np.exp(summary_table.Coefficients)
summary_table
```

```python
summary_table.sort_values('Odds_ratio',ascending = False)
```

```python
# Test the data
reg.score(x_test,y_test)
```

```python
predicted_probability = reg.predict_proba(x_test)
predicted_probability
```

```python
predicted_probability[:,1]
```

```python
x_test
```

```python
prediction = reg.predict(x_test)
prediction.shape
```

```python
Absent_prediction = x_test.copy()
Absent_prediction
```

```python
Absent_prediction['probability'] = predicted_probability[:,1]
Absent_prediction['prediction'] = prediction
Absent_prediction
```
